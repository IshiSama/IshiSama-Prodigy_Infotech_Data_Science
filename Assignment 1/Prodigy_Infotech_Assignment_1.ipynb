```python
import matplotlib.pyplot as plt

# Sample categorical data
genders = ['Male', 'Female', 'Other']
counts = [120, 150, 30]

# Create bar chart
plt.bar(genders, counts)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample continuous data (ages)
ages = [18, 20, 22, 25, 30, 35, 40, 22, 24, 26, 28, 32, 36, 45, 50]

# Create histogram
plt.hist(ages, bins=6)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

plt.show()
```
